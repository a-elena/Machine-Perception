```python
def hessian_points(I, sigma):

    I_xx, I_xy, I_yy = second_derivative(I, sigma)

    det = I_xx * I_yy - I_xy ** 2

    return det / np.max(det)
```

```python
graf_a = imread_gray("data/graf/graf_a.jpg")
det_05 = hessian_points(graf_a, 0.5)
det_1 = hessian_points(graf_a, 1)
det_3 = hessian_points(graf_a, 3)

plt.figure(figsize=(14,8))
plt.subplot(1,3,1)
plt.imshow(det_05, cmap='gray')
plt.title("sigma = 3")
plt.subplot(1,3,2)
plt.imshow(det_1, cmap='gray')
plt.title("sigma = 7")
plt.subplot(1,3,3)
plt.imshow(det_3, cmap='gray')
plt.title("sigma = 13")
plt.show()
```

```python
def new_nonmaxima_suppression_box(I, box_size, thereshold):

    image = np.pad(I, pad_width=box_size//2, mode='constant', constant_values=0)
    result = np.zeros((I.shape))
    rows, cols = image.shape

    for i in range(box_size//2, rows - box_size//2):
        for j in range(box_size//2, cols - box_size//2):
            
            neighbors = image[i - box_size//2:i + box_size//2+1, j - box_size//2:j + box_size//2+1]
            
            if image[i, j] == np.max(neighbors) and image[i, j] > thereshold:
                result[i-box_size//2, j-box_size//2] = image[i, j]
                
    return result
   
```

```python
def draw_large_red_x_on_mask(image, mask, size=300):
    """
    Draws larger red 'X' marks on the image where the mask has non-zero values.
    
    Args:
        image (np.ndarray): The original image (in BGR format).
        mask (np.ndarray): The mask image (grayscale, same dimensions as the original image).
        size (int): The half-length of the 'X' arms (default is 5 pixels).
        
    Returns:
        np.ndarray: The modified image with larger 'X' marks.
    """
    if image.shape[:2] != mask.shape:
        raise ValueError("Mask dimensions must match the image dimensions.")
    
    # Loop through each point in the mask
    for x in range(mask.shape[0]):
        for y in range(mask.shape[1]):
            if mask[x, y] > 0:
                # Draw the 'X' with the specified size
                for i in range(-size, size + 1):
                    # Draw diagonal lines if the indices are valid
                    if 0 <= x + i < mask.shape[0] and 0 <= y + i < mask.shape[1]:
                        image[x + i, y + i] = [0, 0, 255]  # Red color
                    if 0 <= x + i < mask.shape[0] and 0 <= y - i < mask.shape[1]:
                        image[x + i, y - i] = [0, 0, 255]  # Red color
    
    return image

```

```python
# Load the image and mask
image = cv2.imread("data/graf/graf_a.jpg")  # Replace with your image path
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
mask = new_nonmaxima_suppression_box(det_1, 3, 0.1)  # Load mask as grayscale

# Apply the function
result_image = draw_red_x_on_mask(image, mask)

plt.imshow(result_image)
plt.show()
```
